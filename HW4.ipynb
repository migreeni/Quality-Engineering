```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import math
from scipy.special import erfinv
```

### Q 6.10

```python
data_0610 = pd.read_csv('Table_6E5.csv').iloc[:,1:4]
```

```python
data_0610
```

```python
x_bar = [] # average 
r = [] # range of the values
s = [] # standard deviation 

for group in np.array(data_0610):
    x_bar.append(group.mean())
    r.append(group.max() - group.min())
    s.append(np.std(group))


print(statistics.mean(x_bar))
print(statistics.mean(r))

```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(x_bar, linestyle='-', marker='o', color='black')
plt.axhline((statistics.mean(x_bar)+1.023*statistics.mean(r)), color='red', linestyle='--')
plt.axhline((statistics.mean(x_bar)-1.023*statistics.mean(r)), color='red', linestyle='--')
plt.axhline((statistics.mean(x_bar)), color='blue')
plt.title('x-bar Chart', fontsize=30)
plt.xlabel('Sample Number', fontsize=15)
plt.ylabel('Mean', fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(r, linestyle='-', marker='o', color='black')
plt.axhline(2.574*(statistics.mean(r)), color='red', linestyle='dashed')
plt.axhline(0*(statistics.mean(r)), color='red', linestyle='dashed')
plt.axhline((statistics.mean(r)), color='blue')
plt.title('R Chart', fontsize=30)
plt.xlabel('Sample Number', fontsize=15)
plt.ylabel('Range', fontsize=15)
plt.show()
```

### Q 6.11

```python
data_0611 = pd.read_csv('Table_6E6.csv').iloc[:,1:11]
```

```python
x_bar = [] # average 
r = [] # range of the values
s = [] # standard deviation 
var = []

for group in np.array(data_0611):
    x_bar.append(group.mean())
    r.append(group.max() - group.min())
    s.append(np.std(group)*math.sqrt(10)/3)
    var.append((np.std(group)*math.sqrt(10)/3)**2)

print((statistics.mean(x_bar)))
print((1*statistics.mean(s)))
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(var, linestyle='-', marker='o', color='black')
plt.axhline(((statistics.mean(s)**2)*19.02)/9, color='red', linestyle='--')
plt.axhline(((statistics.mean(s)**2)*2.7)/9, color='red', linestyle='--')
plt.axhline((statistics.mean(s)**2), color='blue')
plt.title('S^2 Chart', fontsize=30)
plt.xlabel('Sample Number', fontsize=15)
plt.ylabel('Variance', fontsize=15)
plt.show()


print(((statistics.mean(s)**2)*19.02)/9)
print((statistics.mean(s)**2))
print(((statistics.mean(s)**2)*2.7)/9)
```

## Q 7.3

```python
data_0703 = pd.read_csv('Table_7E1.csv').iloc[:,1:2]

size=50

p_data_0703 = data_0703['Number Nonconforming']/size

```

```python
# Plot p-chart
fig = plt.figure(figsize=(15,8))
plt.plot(p_data_0703, linestyle='-', marker='o', color='black')
plt.axhline(statistics.mean(p_data_0703)+3*(np.sqrt((statistics.mean(p_data_0703)*(1-statistics.mean(p_data_0703)))/(size))), color='red', linestyle ='dashed')
plt.axhline(0, color='red', linestyle ='dashed')
plt.axhline(statistics.mean(p_data_0703), color='blue')

plt.title('p Chart', fontsize=20)
plt.xlabel('Group', fontsize=15)
plt.ylabel('Fraction Defective', fontsize=15)
plt.show()
```

```python
UCL = statistics.mean(p_data_0703)+3*(np.sqrt((statistics.mean(p_data_0703)*(1-statistics.mean(p_data_0703)))/(size)))
LCL = 0
out_list = []
i = 0
control = True
for group in p_data_0703:
    if group > UCL or group < LCL:
        print('Out of control limits -> Day {}'.format(i+1))
        out_list.append(i)
        control = False
    i += 1
if control == True:
    print('All groups are within control limits.')
```

```python
new_p_data = p_data_0703.drop(out_list).reset_index(drop = True)
```

```python
new_p_data
```

```python
fig = plt.figure(figsize=(15,8))
plt.plot(new_p_data, linestyle='-', marker='o', color='black')
plt.axhline(statistics.mean(new_p_data)+3*(np.sqrt((statistics.mean(new_p_data)*(1-statistics.mean(new_p_data)))/(size))), 
            color='red', linestyle ='dashed')
plt.axhline(0,color='red', linestyle ='dashed')
plt.axhline(statistics.mean(new_p_data), color='blue')

plt.title('Revised p Chart', fontsize=20)
plt.xlabel('Group', fontsize=15)
plt.ylabel('Fraction Defective', fontsize=15)
plt.show()

print(statistics.mean(new_p_data)+3*(np.sqrt((statistics.mean(new_p_data)*(1-statistics.mean(new_p_data)))/(size))))
print(statistics.mean(new_p_data)-3*(np.sqrt((statistics.mean(new_p_data)*(1-statistics.mean(new_p_data)))/(size))))
```

```python
data_0723 = pd.read_csv('Table_7E9.csv').iloc[:,2:3]

c_data = data_0723['Number Nonconforming']
```

```python
fig = plt.figure(figsize=(15,8))
plt.plot(c_data, linestyle='-', marker='o', color='black')
plt.axhline(statistics.mean(c_data)+3*np.sqrt(statistics.mean(c_data)), 
            color='red', linestyle='dashed')
plt.axhline(statistics.mean(c_data)-3*np.sqrt(statistics.mean(c_data)), 
            color='red', linestyle='dashed')
plt.axhline(statistics.mean(c_data), color='blue')

plt.title('c Chart', fontsize=20)
plt.xlabel('Group', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
UCL = statistics.mean(c_data)+3*np.sqrt(statistics.mean(c_data))
LCL = statistics.mean(c_data)-3*np.sqrt(statistics.mean(c_data))
out_list = []
i = 0
control = True
for group in c_data:
    if group > UCL or group < LCL:
        print('Out of control limits -> Sample number {}'.format(i+1))
        out_list.append(i)
        control = False
    i += 1
if control == True:
    print('All groups are within control limits.')
```

```python
out_list
```

```python
new_c_data=(c_data.drop(out_list)).reset_index(drop = True)
new_c_data
```

```python
fig = plt.figure(figsize=(15,8))
plt.plot(new_c_data, linestyle='-', marker='o', color='black')
plt.axhline(statistics.mean(new_c_data)+3*np.sqrt(statistics.mean(new_c_data)), 
            color='red', linestyle='dashed')
plt.axhline(statistics.mean(new_c_data)-3*np.sqrt(statistics.mean(new_c_data)), 
            color='red', linestyle='dashed')
plt.axhline(statistics.mean(new_c_data), color='blue')

plt.title('revised c Chart', fontsize=20)
plt.xlabel('Group', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()

print(statistics.mean(new_c_data)+3*np.sqrt(statistics.mean(new_c_data)))
print(statistics.mean(new_c_data)-3*np.sqrt(statistics.mean(new_c_data)))
print(statistics.mean(new_c_data))
```
